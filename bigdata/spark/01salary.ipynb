# count the no of employees in particular salary

# The file fields are id,name, salary

```python
import pyspark
```

```python
from pyspark import SparkContext
```

```python
sc = SparkContext()
```

```python
input = sc.textFile("/user/gl_faculty_greatlearning/datasets/emp.txt")
```

```python
input
```

```python
input.top(7)
```

```python
salaries = input.map(lambda x: x.split(',')[2])
```

```python
salaries.top(2)
```

```python
results = salaries.countByValue()
```

```python
results
```

```python
for sal, count in results.items():
	print(sal + " " + str(count))
```

```python
sc.stop()
```
