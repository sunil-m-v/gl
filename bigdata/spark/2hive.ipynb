# Spark SQL Hive integration

```python
spark = SparkSession.builder.config("spark.some.config.option", "some-value").embedHiveSupport().getOrCreate()
```

```python
spark.sql("CREATE TABLE IF NOT EXISTS src(key INT, value STRING)")
```

```python
spark.sql("LOAD DATA INPATH 'hdfs:///user/gl_faculty_greatlearning/spark_recording/kv1.txt' INTO TABLE src")
```

```python
#Queries are expressed in HiveQL
spark.sql("SELECT * FROM src").show()
```

```python
#Aggregation queries are also supported
spark.sql("SELECT COUNT(*) FROM src").show()
```

```python
#The results of SQL queries are themeselves DataFrames and support all normal functions.
sqlDF = spark.sql("SELECT key, value FROM src WHERE key < 10 ORDER BY key")
```

```python
# The items in DataFramesare of type Row, which allows you to access each column.
stringDS = sqlDF.rdd.map(lambda row: "Key: %d, Value: %s" %(row.key, row.value))
```
