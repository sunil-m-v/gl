```python
# first line is empty and header is at second row. 1st line is 0th row and 2nd line is 1st row
liga=pd.read_csv('C:\\Users\\VMuvvala\\Desktop\\GL\\Laliga.csv',header=1)
```

```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
liga=liga.replace(to_replace="-",value='0',regex=False)
#df.replace(to_replace ='[nN]ew', value = 'New_', regex = True)
```

```python
#Print all the teams which have started playing between 1930-1980
# started playing between 1930 and 1980
# if the value is 1930-1931 then it is taken lower boundary as 1930 and upper boundary as 1931
# if it is 1930 then lower and upper boundary are same
print(liga[(liga['Debut'].apply(lambda x:int(x.split('-')[0])) >= 1930) & 
           (liga['Debut'].apply(lambda x: (int(x.split('-')[0]) + 1) if len(x.split('-')) > 1 else int(x.split('-')[0]))  <= 1980)]['Team'].unique())
```

```python
#Print the list of teams which came Top 5 in terms of points
# first convert Points column to int.
liga['Points']=liga['Points'].astype(int)
liga_sorted=liga.sort_values(by='Points', inplace=False,ascending=False)
print(liga_sorted['Team'].head(5))
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, 
#find the team which has maximum and minimum goal difference.
def Goal_diff_count(x,y):
    return x-y

liga['Goal_diff_count']=Goal_diff_count(liga['GoalsFor'].astype(int),liga['GoalsAgainst'].astype(int))
team_max_goal_cnt=liga[liga['Goal_diff_count']==liga['Goal_diff_count'].max()]['Team']
team_min_goal_cnt=liga[liga['Goal_diff_count']==liga['Goal_diff_count'].min()]['Team']
print(team_max_goal_cnt)
print(team_min_goal_cnt)
```

```python
#Create a new column with name “Winning Percent” and append it to the data set
# fill Nan values resulting from 0/0 with 0
# Print the top 5 teams which has the highest Winning percentage
liga['Winning Percent']=(((liga['GamesWon'].astype(int))*100)/liga['GamesPlayed'].astype(int)).fillna(0)
ligs_sort_win=liga.sort_values('Winning Percent',inplace=False,ascending=False)
print(ligs_sort_win.head(5)['Team'])
```

```python
# Group teams based on their “Best position” and print the sum of their points for all positions
liga.groupby('BestPosition').sum().sort_values(by='Points',ascending=False)['Points']
```
